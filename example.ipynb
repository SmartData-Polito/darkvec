```python
from idarkvec.utils import *
from idarkvec.models.word2vec import Word2Vec
from idarkvec.preprocessing import CorpusExtractor

TRACES = 'data/raw_traces'
WORD2VEC = 'data/word2vec'
darknet = 'darknet01'
start, end, cnt = '20210501', '20210502', 3

while start != get_next_day(end):
    current_day = start
    while current_day != get_next_day(end):
        corpus = CorpusExtractor(trace_path=f'{TRACES}/{darknet}/{darknet}', day=current_day, 
                                 min_freq=5).from_darknet(top_ports=2500, verbose=True)
        if current_day == start:
            mname = f'training_{cnt}_s_{darknet}_{current_day}_d_{darknet}_{current_day}'
            word2vec = Word2Vec(c=5, e=200, destination=f'{WORD2VEC}/{mname}')
            word2vec.train(corpus, save=True)
        else:
            sname = mname
            mname = f'training_{cnt}_s_{darknet}_{prev_day}_d_{darknet}_{current_day}'
            word2vec = Word2Vec(c=5, e=200, source=f'{WORD2VEC}/{sname}', destination=f'{WORD2VEC}/{mname}')
            word2vec.update(corpus, save=True)

        prev_day = current_day
        current_day = get_next_day(current_day)

    start = get_next_day(start)
    cnt -= 1
```
